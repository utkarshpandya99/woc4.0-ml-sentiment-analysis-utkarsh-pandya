```python
finaldata['tweet']=finaldata['tweet'].str.lower()
finaldata['tweet']
```

```python
import regex as re
def remove_url(x):
    return re.sub('((www.[^s]+)|(https?://[^s]+))',' ',x)
finaldata['tweet'] = finaldata['tweet'].apply(lambda x: remove_url(x))
finaldata['tweet']
```

```python
import string
punct = string.punctuation
def cleaning_punctuations(x):
    trans = str.maketrans('', '', punct)
    return x.translate(trans)
finaldata['tweet']= finaldata['tweet'].apply(lambda x: cleaning_punctuations(x))
finaldata['tweet']
```

```python
def remove_numbers(x):
    return re.sub('[0-9]+', '', x)
finaldata['tweet'] = finaldata['tweet'].apply(lambda x: remove_numbers(x))
finaldata['tweet']
```

```python
def remove_spaces(x):
  return " ".join(x.split())
finaldata['tweet'] = finaldata['tweet'].apply(lambda x:remove_spaces(x))
finaldata['tweet']
```

```python
import nltk
nltk.download('punkt')
def token(x):
  return nltk.word_tokenize(x)
finaldata['tweet'] = finaldata['tweet'].apply(lambda x:token(x))
finaldata['tweet'].head()
```

```python
stm = nltk.PorterStemmer()
def stemming(x):
    text = [stm.stem(i) for i in x]
    return text
finaldata['stem_tweet'] = finaldata['tweet'].apply(lambda x: stemming(x))
finaldata['stem_tweet']
```

```python
x_train = vectoriser.transform(x_train)
x_test  = vectoriser.transform(x_test)
```
